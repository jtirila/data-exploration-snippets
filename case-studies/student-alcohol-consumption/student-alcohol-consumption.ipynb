# Analyzing the student alcohol consumption regression data

See the private wiki's StudentAlcoholConsumption page for details. 

For now, just performing some exploration on the data and finding out about what one can do with it using SciPy. 

This is very much work in progress. 

```python
%matplotlib inline
import csv
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

with open('student-mat.csv', newline='') as csvfile:
    sniffer = csv.Sniffer()
    # Detect delimiter, quotechar etc. automatically
    dialect = sniffer.sniff(csvfile.read(1024))
    # Go back to beginning
    csvfile.seek(0)
    csvreader = csv.reader(csvfile, dialect)
    data = list(csvreader)
    data.pop(0)
    indices = range(len(data))
    dalc = []
    walc = []
    stime = []
    failures = []
    
    for datarow in data:
        stime.append(datarow[13])
        failures.append(datarow[14])
        dalc.append(datarow[26])
        walc.append(datarow[27])
    
    # print(data)
    # print(len(data))
    # print(dalc)
    # print(len(dalc))
    # print(walc)
    # print(len(walc))
    # print(len(indices))
    
    plt.scatter(indices, dalc)
    plt.show()
    
    plt.scatter(indices, walc)
    plt.show()

    # So yeah, I realize the rest of the plots really make no sense, the data is so descreet and I should
    # rather be interested in the number of points at each intersection. No time for that today. 

    plt.scatter(stime, dalc)
    plt.show()
    plt.scatter(failures, dalc)
    plt.show()
    plt.scatter(stime, walc)
    plt.show()
    plt.scatter(failures, walc)
    plt.show()

    
```
